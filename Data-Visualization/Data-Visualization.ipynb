## Matplotlib Kütüphanesi

```python
# Matplotlib kütüphanesini import edelim

import matplotlib.pyplot as plt

%matplotlib inline  
# grafikleri satır içinde göster
```

```python
import numpy as np

# Grafik çizmeden önce rastgele veriler oluşturalım

x = np.arange(20)               # 0-19 arası sayılardan oluşan bir dizi oluştur
y = np.random.normal(10, 1, 20) # Merkezi 10 olan 1 standart sapmalı 20 adet normal dağılımlı rastgele sayı oluştur
z = np.random.normal(10, 2, 20) # Merkezi 10 olan 2 standart sapmalı 20 adet normal dağılımlı rastgele sayı oluştur

print(x)
print(y)
print(z)
```

#### Çizgi Grafiği (Plot)

```python
plt.plot(x,y,z)
plt.xlabel(" X Değerleri")
plt.ylabel(" Y ve Z Değerleri")
plt.title("X'in Y ve Z'ye göre değerleri")
```

#### Dağılım Grafiği (Scatter Plot)

```python
plt.scatter(x,y)
plt.title('Scatter plot')
```

#### Histogram Grafiği

```python
plt.hist(y);
```

#### Toplu Grafik Gösterimi

```python
fig=plt.figure(figsize=(10,3))

ax=fig.add_subplot(131)  # 1 satır, 3 sütunlu tablonun 1. hücresi
ax.plot(x,y)

bx=fig.add_subplot(132)  # 1 satır, 3 sütunlu tablonun 2. hücresi
bx.scatter(x,z, c="red", marker="d")

bx=fig.add_subplot(133)  # 1 satır, 3 sütunlu tablonun 3. hücresi
bx.hist(y);
```

## Pandas Grafikleri

```python
import pandas as pd

df = pd.read_csv("train.csv")
df
```

#### Age ve Fare alanına göre dağılım grafiği

```python
df.plot.scatter(x='Age', y='Fare', title='Scatter plot')
```

#### Fare ve Age alanlarını ayrı ayrı çizgi grafiği

```python
df[['Fare', 'Age']].plot.line(title='Line chart')
```

#### Age alanı verilerinin Histogram grafiği

```python
df['Age'].plot.hist()
```

#### Bar Plot

```python
# Dikey

df['Pclass'].value_counts().sort_index().plot.bar()
```

```python
# Yatay

df['Pclass'].value_counts().sort_index().plot.barh()
```

#### Box plot

```python
df['Age'].plot.box()
```

##  Seaborn Grafikleri

```python
import seaborn as sns
```

#### Scatter Plot

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
```

#### Line Plot

```python
sns.lineplot(data=df[['Age','Fare']])
```

#### Histogram / Distribution Plot

```python
sns.displot(df['Age'], bins=10, kde=True) # Bins: bar sayısı
```

#### Bar Chart/ Count Plot

```python
sns.countplot(x=df['Embarked'])
```

#### Box Plot

```python
sns.boxplot(x=df['Age'], y=df['Sex'])
```

#### Korelasyon - Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

#### Pair Plot 
Survived alanına göre bütün sütunları bütün sütunlarla karşılaştır

```python
sns.pairplot(df, hue='Survived')
```
